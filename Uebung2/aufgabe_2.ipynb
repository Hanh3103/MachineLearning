```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) Pearson
import math

def pearson_corr(data_x , data_y):
    mean_x = sum(data_x)/ len(data_x)
    mean_y = sum(data_y)/ len(data_y)

    substr_mean_x = [i - mean_x for i in data_x]
    substr_mean_y = [i - mean_y for i in data_y]
    
    x_multiple_y = [a * b for a, b in list(zip(substr_mean_x, substr_mean_y))]
    print(sum(x_multiple_y))
    data_x_squared = [i * i for i in substr_mean_x]
    print(sum(data_x_squared))
    data_y_squared = [i * i for i in substr_mean_y]
    print(sum(data_y_squared))

    return sum(x_multiple_y)/(math.sqrt(sum(data_x_squared)*sum(data_y_squared)))

X = [12,11,13,13, 9,10,10,13, 5,10,10,13,10,10, 5, 8, 9, 8, 8, 9, 9,10,11, 5,12]
Y = [11,10,10,10, 9,13,10,11, 6, 7,13,14,14,11,11,10,10, 7, 8,12,11,11, 8, 7,13]

print(pearson_corr(X, Y))
```
