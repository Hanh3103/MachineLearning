```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
array = [4,2,5,6,1,6,8,3,4,9]
mean = np.mean(array)
print(mean)
```

```python
from scipy.stats import stats
```

```python
median = np.median(array)
print(median)
mode = stats.mode(array)
print(mode)
print(mode[0])
print(mode[1])
```

```python
variance = np.var(array)
print(variance)
```

```python
#Standard deviation of data
sd = np.std(array)
print(sd)
```

```python
# 10% of the data value smaller als 1.9
percentil_10 = np.percentile(array, 10)
print(percentil_10)
percentil_50 = np.percentile(array, 50)
print(percentil_50)
percentil_100 = np.percentile(array, 100)
print(percentil_100)
```

```python
#generate a uniformly distributed data set as an array
#generating 500 random values ranging from 60 to 120
#heart_rates = np.random.uniform(60,120,500)
heart_rates = np.random.randint(60,120+1,500)
print(heart_rates)
```

```python
# plot a historgram with heart rate over x axis and no of occurences over y axis
# creating a historgram with 60 bars in the plot
plt.hist(heart_rates,60)
plt.show()
```

```python
#generate a normally distributed data set as an array
#generating 500 random values 
#arguments are mean,SD,no of samples required
heart_rates = np.random.normal(72,5,500)
print(heart_rates)
```

```python
# plot a historgram with heart rate over x axis and no of occurences over y axis
# creating a historgram with 60 bars in the plot
plt.hist(heart_rates,60)
plt.show()
```

```python
# finding the z score of the heart rate array
z_scores = stats.zscore(heart_rates)
print(z_scores)
```
