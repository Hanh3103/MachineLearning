```python
import generator
import pandas as pd
data1 = pd.DataFrame(generator.data_generator(-0.1,0.7,100))
data2 = pd.DataFrame(generator.data_generator(0.1,2,100))
data3 = pd.DataFrame(generator.data_generator(0.75,-0.2,100))
data4 = pd.DataFrame(generator.data_generator(1,-1,100))
```

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def visualize(data):

    # Separate data points by class
    df = pd.DataFrame(data)
    df1 = data[data.loc[:,2] >= 0]
    df2 = data[data.loc[:,2] <= 0]


    # Plot data points and hyperplane
    plt.scatter(df1.loc[:, 0], df1.loc[:, 1], color='r')
    plt.scatter(df2.loc[:, 0], df2.loc[:, 1], color='b')

    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    X = df[[0, 1]].to_numpy()
    y = df[2].to_numpy()
    clf = SVC(C = 1, kernel = 'linear')
    clf.fit(X,y)
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    X1, X2 = np.meshgrid(Y, X)
    fvalues = np.vstack([X.ravel(), Y.ravel()]).T

    F = clf.decision_function(fvalues).reshape(X.shape)

    ax.contour(X, Y, F, levels=[-1, 0, 1])

    # Add labels and show plot
    plt.xlabel('w1')
    plt.ylabel('w2')
    plt.title('SVM Hyperplane and Edges')
    plt.show()
```

```python
visualize(data1)
visualize(data2)
visualize(data3)
visualize(data4)
```
